**handling null columns**

```python
# handle CustomerID column
df['CustomerID']=df['CustomerID'].fillna(df['CustomerID'].median())
df['CustomerID'].isnull().sum()
```

```python
# handle Description column
df['Description']=df['Description'].fillna(0)
df['Description'].isnull().sum()
```

```python
df.drop(columns='UnitPrice',inplace=True)
df.columns
```

```python
(df.isnull().sum()/df.shape[0])*100
```

```python
# handle null values in (invoice - stockcode)
df['InvoiceNo'].fillna(0,inplace=True)
df['InvoiceNo'].isnull().sum()
```

```python
print(df['StockCode'].mean())
print(df['StockCode'].median())
```

```python
df['StockCode']=df['StockCode'].fillna(df['StockCode'].median())
df['StockCode'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Country'],inplace=True)
df.isnull().sum()
```
